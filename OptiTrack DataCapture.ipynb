```python
#with date

import requests
import json
import datetime
from threading import Thread
from time import sleep
import csv
import sqlite3
from activity import Activity
import time

#create a new db and connect or simply connect if already exists

conn = sqlite3.connect('pilot_test.db')

c = conn.cursor()

#please comment c.execute section once the db is created

c.execute("""CREATE TABLE activities (
            date text,
            sourceId integer,
            posx integer,
            posy integer,
            posz integer,
            orx integer,
            ory integer,
            orz integer
            )""")

done = False

def listen_for_enter_key_press():
    global done
    input()
    done = True
    
#api for data capture 
    
url = "https://thor.evl.uic.edu:4000/mocap"
#content = requests.get(url).content
#dataset = json.loads(content)

# print(dataset)

dataframe=list()

listener = Thread(target=listen_for_enter_key_press)
listener.start()
count = 0

def insert_act(act):
    with conn:
        c.execute("INSERT INTO activities VALUES ( :date, :sourceId, :posx, :posy, :posz, :orx, :ory, :orz)", 
              {'date': act.date, 'sourceId': act.sourceId, 'posx': act.posx, 'posy': act.posy, 'posz': act.posz, 'orx': act.orx, 'ory': act.ory, 'orz': act.orz})
        conn.commit()
        
while not done:
    content = requests.get(url).content
    dataset = json.loads(content)
    
    for row in dataset:
        date = time.strftime("%H:%M:%S", time.gmtime())
        sourceId = row['sourceId']
        posx = row['posx']
        posy = row['posy']
        posz = row['posz']
        orx = row['orx']
        ory = row['ory']
        orz = row['orz']
        data = (row['timestamp'], sourceId, posx, posy, posz, orx, ory,orz)
        act = Activity(time.strftime("%H:%M:%S", time.gmtime()), sourceId, posx, posy, posz, orx, ory, orz)
        insert_act(act)
        conn.commit()
        dataframe.append(data)
        sleep(0.5)
        count += 1
        print(count)
# #     print(row['sourceIP'])

listener.join()
    
print(dataframe)

conn.close()
```
