```python
#Script for database creation and recording data with time in seconds from start

import requests
import json
import datetime
from threading import Thread
from time import sleep
import csv
import sqlite3
from activity import Activity
import time

#create a new db and connect or simply connect if already exists

conn = sqlite3.connect('pilot_time6.txt')

c = conn.cursor()

#please comment c.execute section once the db is created

# c.execute("""CREATE TABLE activities (
#             sourceId integer,
#             posx integer,
#             posy integer,
#             posz integer,
#             orw integer,
#             orx integer,
#             ory integer,
#             orz integer,
#             time integer
#             )""")

done = False

def listen_for_enter_key_press():
    global done
    input()
    done = True
    
#api for data capture 
    
url = "https://thor.evl.uic.edu:4000/mocap"
#content = requests.get(url).content
#dataset = json.loads(content)

# print(dataset)

dataframe=list()

listener = Thread(target=listen_for_enter_key_press)
listener.start()
count = 0

def insert_act(act):
    with conn:
        c.execute("INSERT INTO activities VALUES ( :sourceId, :posx, :posy, :posz, :orw, :orx, :ory, :orz, :time)", 
              {'sourceId': act.sourceId, 'posx': act.posx, 'posy': act.posy, 'posz': act.posz, 'orw': act.orw, 'orx': act.orx, 'ory': act.ory, 'orz': act.orz, 'time': act.time})
        conn.commit()

starttime = time.time()
        
while not done:
    content = requests.get(url).content
    dataset = json.loads(content)
    
    for row in dataset:
        #date = time.strftime("%H:%M:%S", time.gmtime())
        sourceId = row['sourceId']
        posx = row['posx']
        posy = row['posy']
        posz = row['posz']
        orw = row['orw']
        orx = row['orx']
        ory = row['ory']
        orz = row['orz']
        totaltime = round((time.time() - starttime), 2)
#         print(totaltime)
        data = (sourceId, posx, posy, posz, orw, orx, ory,orz,totaltime)
        act = Activity(sourceId, posx, posy, posz, orw, orx, ory, orz, totaltime)
        insert_act(act)
        conn.commit()
        dataframe.append(data)
        sleep(0.1)
        count += 1
#         print(count)
# #     print(row['sourceIP'])

listener.join()
    
print(dataframe)



#conn.close()


```

```python
# script to convert .db to .txt

import sqlite3
import pandas as pd

path = 'pilot_time5.db'
conn = sqlite3.connect(path)
header = []
c = conn.cursor()
for column in c.execute('PRAGMA table_info("activities")'):
    header.append(column[1])
    
#Insert each data into dataframe and save as csv
df = pd.DataFrame(columns=header)
for raw in c.execute('SELECT * FROM activities ORDER BY sourceId'):
    series = pd.Series(list(raw), index=df.columns)
    df = df.append(series, ignore_index=True)
#df.to_csv('pilot_test.txt')
df.to_csv(r'pilot_time5.txt', header=None, index=None, sep=',', mode='a')
```
