```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import TextFormatter
from youtube_transcript_api.formatters import JSONFormatter

transcript = YouTubeTranscriptApi.get_transcript("OlwRrC1Gd3M", languages=['en'])

formatter1 = TextFormatter()


# .format_transcript(transcript) turns the transcript into a JSON string.
text_formatted = formatter1.format_transcript(transcript)


# Now we can write it out to a file.
with open('firsttest_8minutes.txt', 'w', encoding='utf-8') as text_file:
    text_file.write(text_formatted)
    
formatter2 = JSONFormatter()


# .format_transcript(transcript) turns the transcript into a JSON string.
json_formatted = formatter2.format_transcript(transcript)


# Now we can write it out to a file.
with open('firsttest_8minutes.json', 'w', encoding='utf-8') as json_file:
    json_file.write(json_formatted)
```

```python
import json
import random





convo_list = []

#contact_record = ""

def printrecord(contact_record):
    ids = ["001", "002"]
    xvalues = ["0", "0.8", "0.5", "-0.8", "-0.5"]
    yvalue = "-1.5"
    zvalues = ["2", "1.5", "-1.5", "2.5"]
    zvalue = "2"
    prevtime = 0
    with open('8minutes.txt', 'w') as f:
        for convo in contact_record:
            for k, v in convo.items():
                if(k == "text"):
                    text = v
                if(k == "start"):
                    time = v - prevtime
                    prevtime = v
            newline = random.choice(ids) +" "+ random.choice(xvalues) +" "+ yvalue +" "+ zvalue +" " +str(round(time, 2)) +", " + text + "\n"
            print(newline)
            f.write(newline)
    f.close()
                
        
        


def read_contact(filename):
    with open(filename, 'r') as f:
        contact_record = json.load(f)
        #print(contact_record)
        printrecord(contact_record)
    #f.close()

read_contact("firsttest_8minutes.json")




```
